# üìÇ3.DATA MERGING

### Merge season info into deliveries dataset using 'match_id'

```python
ipl= deliveries.merge(
    matches,
    left_on='match_id',  # Column from `deliveries` to merge on
    right_on='id',       # Column from `matches` to merge on
    how='left'           # Left join: keeps all rows from `deliveries`, adds matching rows from `matches`
)
```
